## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata_df = pd.read_csv(mouse_metadata_path)
study_results_df = pd.read_csv(study_results_path)

# Combine the data into a single dataset
raw_merged_df = study_results_df.merge(mouse_metadata_df, how='left', on=['Mouse ID'])

# Display the data table for preview
print('\nMerged Data Table of Mouse Metadata & Study Results')
raw_merged_df
```

```python
# Checking the number of mice (with duplicates)
all_miceCount = study_results_df['Mouse ID'].count()

print(f'\nCount of All Mice (Including Duplicates): {all_miceCount}')
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicated_mice_df = study_results_df.groupby(['Mouse ID','Timepoint'])['Tumor Volume (mm3)'].count().to_frame('Check Duplicates').reset_index()
duplicated_mice_df = duplicated_mice_df[duplicated_mice_df['Check Duplicates'] > 1]

# output duplicate mice by ID number & Timepoint
print('\nData Table of Duplicate Mouse ID & Timepoint')
#duplicate_miceID_df

duplicated_mice_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_allData_df = raw_merged_df[(raw_merged_df['Mouse ID'] == "g989") & 
                                          ((raw_merged_df['Timepoint'] == 0)  |
                                           (raw_merged_df['Timepoint'] == 5)  |
                                           (raw_merged_df['Timepoint'] == 10) |
                                           (raw_merged_df['Timepoint'] == 15) |
                                           (raw_merged_df['Timepoint'] == 20))]


# & study_results_df['Timepoint'] == 0
# df.drop(df[df['Age'] < 25].index, inplace = True) 

print('\nData Table of Duplicate Mouse ID & Timepoint, Showing All Data Columns')
duplicate_mice_allData_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
dropIndex = duplicate_mice_allData_df.index.values.tolist()

noDuplicates_merged_df = raw_merged_df.drop(dropIndex)

# Display the merged data table for preview
print('\nMerged Data Table With Dropped Duplicate Mice By ID & Timepoint')
noDuplicates_merged_df
```

```python
# Checking the number of mice in the clean DataFrame.
noDuplicates_miceCount = noDuplicates_merged_df['Mouse ID'].nunique()

print(f'\nCount of Unique Mice: {noDuplicates_miceCount}')
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

groupbyDrugs_merged_df = noDuplicates_merged_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen')
summaryStats_groupbyDrugs_merged_df = pd.DataFrame(groupbyDrugs_merged_df.describe())

# Add calculated columns for variance and standard error
summaryStats_groupbyDrugs_merged_df[('Tumor Volume (mm3)', 'var')] = summaryStats_groupbyDrugs_merged_df[('Tumor Volume (mm3)', 'std')]**2
summaryStats_groupbyDrugs_merged_df[('Tumor Volume (mm3)', 'SEM')] = summaryStats_groupbyDrugs_merged_df[('Tumor Volume (mm3)', 'std')] / summaryStats_groupbyDrugs_merged_df[('Tumor Volume (mm3)', 'count')]**0.5

# Remove unnecessary columns (25%, 75%, and max)
updated_summaryStats_groupbyDrugs_df = summaryStats_groupbyDrugs_merged_df.drop(columns = [('Tumor Volume (mm3)', 'count'),
                                                                                           ('Tumor Volume (mm3)', 'min'),
                                                                                           ('Tumor Volume (mm3)', '25%'),
                                                                                           ('Tumor Volume (mm3)', '75%'),
                                                                                           ('Tumor Volume (mm3)', 'max')])

# updated_summaryStats_groupbyDrugs_df.columns.set_levels(['Mean','Std. Dev','Median','Variance','SEM'],level=1,inplace=True)

print('\nSummary Statistics of Tumor Volume (mm3) By Drug Regimens (Using Describe Method)')
updated_summaryStats_groupbyDrugs_df = updated_summaryStats_groupbyDrugs_df.round(1)
updated_summaryStats_groupbyDrugs_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
agg_df = noDuplicates_merged_df[['Drug Regimen','Tumor Volume (mm3)']].groupby('Drug Regimen').agg(['mean','median','std','var','sem'])

# outputting summary statistics table using aggregate method
print('\nSummary Statistics of Tumor Volume (mm3) By Drug Regimens (Using Aggregate Method)')
agg_df = agg_df.round(1)
agg_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
maxTime_by_mice = noDuplicates_merged_df.groupby('Mouse ID')['Timepoint'].max().to_frame(name = 'Max Time')

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_vol_df = maxTime_by_mice.merge(noDuplicates_merged_df, how='inner', left_on=["Mouse ID", "Max Time"], right_on=["Mouse ID","Timepoint"])

# Filter to just four drugs: Capomulin, Ramicane, Infubinol, and Ceftamin
tumor_vol_df = tumor_vol_df[tumor_vol_df['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# output tumor volume
print('\nData Frame of Final Timepoint Tumor Volumes & Other Fields For Each Mouse')
tumor_vol_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = {}

# Create empty dataframe to output quartile calculations
output_quartiles_df = pd.DataFrame(index = ['Lower Bound','Quartile 1','Quartile 3','Upper Bound', 'IQR'], columns = treatment_list)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatment_list:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    filtered_tumor_vol_df = tumor_vol_df[tumor_vol_df['Drug Regimen'] == treatment]

    # add subset (tumor volume)
    
    tumor_vol_list = filtered_tumor_vol_df['Tumor Volume (mm3)'].to_list() 
    tumor_vol[treatment] = tumor_vol_list
    
    # calculate quariles & IQRs
    q1 = filtered_tumor_vol_df['Tumor Volume (mm3)'].quantile(0.25).round(1)
    q3 = filtered_tumor_vol_df['Tumor Volume (mm3)'].quantile(0.75).round(1)
    iqr = round(q3 - q1,1)
    lower_bound = round(q1 - 1.5*iqr,1)
    upper_bound = round(q1 + 1.5*iqr,1)
    
    tumor_vol[treatment + ' Quartiles'] = [lower_bound, q1, q3, upper_bound, iqr]
    update_series = pd.Series(tumor_vol[treatment + ' Quartiles'], name=treatment, index=['Lower Bound','Quartile 1','Quartile 3','Upper Bound', 'IQR'])
    update_series = update_series.round(1)
    output_quartiles_df[treatment].update(update_series)
    
    # Determine outliers using upper and lower bounds
    outlier_df = filtered_tumor_vol_df['Tumor Volume (mm3)'][(filtered_tumor_vol_df['Tumor Volume (mm3)'] < lower_bound) | (filtered_tumor_vol_df['Tumor Volume (mm3)'] > upper_bound)]
    outlier_df = outlier_df.round(1)
    if outlier_df.shape[0] > 0:
        print(f'{treatment} Outliers (Tumor Volume, mm3): {outlier_df.to_list()}')
    else:
        print(f'{treatment} No Outliers')

# Output box plot quartiles
print("\nBox Plot Quartiles: Final Tumor Volume (mm3) by Drug Regimen")
output_quartiles_df
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

df = tumor_vol_df.boxplot(by='Drug Regimen', column=['Tumor Volume (mm3)'], grid=False)
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse b128 treated with Capomulin

# Filter data table to just mouse b128, timepoint, and tumor volume
og_cap_df = noDuplicates_merged_df[(noDuplicates_merged_df['Drug Regimen'] == 'Capomulin') & (noDuplicates_merged_df['Mouse ID'] == 'b128')]
cap_df = og_cap_df[['Mouse ID','Timepoint','Tumor Volume (mm3)']]

x_axis = cap_df['Timepoint'].to_list()
y_axis = cap_df['Tumor Volume (mm3)'].to_list()

# Creating the plot 
line_df = pd.DataFrame({'Mouse b128': y_axis}, index = x_axis)
line_plot = line_df.plot.line()
plt.title("Capomulin Treatment: Tumor Volume Over Time For Mouse ID b128")
plt.xlabel('Days')
plt.ylabel('Tumor Volume (mm3)')
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# Filter data table to just tumor volume, and mouse weight
cap_weight_df = noDuplicates_merged_df[['Tumor Volume (mm3)','Weight (g)']]
cap_weight_df = pd.DataFrame(cap_weight_df.groupby('Weight (g)').mean())

scatter_df = cap_weight_df.reset_index()

# Creating the plot 
line_plot = scatter_df.plot.scatter(x='Weight (g)',y='Tumor Volume (mm3)',c='DarkBlue')
plt.title("Capumolin Treatment: Tumor Volume vs. Average Mouse Weight (g)")
plt.xlabel('Average Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regime

x = scatter_df['Weight (g)'].to_list()
y = scatter_df['Tumor Volume (mm3)'].to_list()
slope, intercept, r_value, p_value, std_err = st.linregress(x,y)

print(f'\nCorrelation Coefficient, r: {r_value.round(2)}')

# calculate line-of-best-fit values
scatter_df['line_y'] = pd.Series([slope*element + intercept for element in x],index = scatter_df.index)

scatter_df

# Creating the plot 
line_plot = scatter_df.plot.scatter(x='Weight (g)',y='Tumor Volume (mm3)',c='DarkBlue')
plt.title("Capumolin Treatment: Tumor Volume vs. Average Mouse Weight (g)")
plt.xlabel('Average Mouse Weight (g)')
plt.ylabel('Tumor Volume (mm3)')

# plotting line-of-best-fit
plt.plot('Weight (g)', 'line_y', data=scatter_df, marker='', color='red', linewidth=1)
```
